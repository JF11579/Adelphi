```python
import pandas as pd
import numpy
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn
import calendar
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split #Split data in testing and training
from sklearn import metrics
from sklearn import preprocessing
# use automatically configured the lasso regression algorithm
from numpy import arange
from pandas import read_csv
from sklearn.linear_model import LassoCV
from sklearn.model_selection import RepeatedKFold
```

```python
data= pd.read_csv('/content/Porcelana_OP_adelphi_2021_SCALED.csv')
data.head(2)
```

```python
import datetime as dt
data['date'] = pd.to_datetime(data['date'])
data['date']=data['date'].map(dt.datetime.toordinal)
```

```python
data.head(3)
```

```python
#Split the data

#X, y = data[:, :-1], data[:, -1]
X= data.loc[:, data.columns != 'Quantity']
y = data.iloc[:,1]  # <- Quantity isolated
#X = data.iloc[:, :2]
X.head(3)
```

```python
y.head(3)
```

```python
import datetime as dt
data['date'] = pd.to_datetime(data['date'])
data['date']=data['date'].map(dt.datetime.toordinal)
```

```python
dataframe = data
```

```python
data = dataframe.values
```

```python
# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
```

```python
# define model
model = LassoCV(alphas=arange(0, 1, 0.01), cv=cv, n_jobs=-1)
```

```python
# fit model
model.fit(X, y)
```

```python
def get_coefficients(model,X):

    predictors_model = pd.DataFrame({'label':X.columns,'coeff':model.coef_})
    return predictors_model

get_coefficients(model,X)
```

```python
coef_df = pd.DataFrame(get_coefficients(model,X))
```

```python
coef_df.shape

```

Let us drop zero values

```python
#coef_df = coef_df[ > 0]
coef_df = coef_df[coef_df['coeff'] != 0]
```
