```r
#your turn # 2 p 338
# step 1 : Get z score using pnorm at %5

# step 2 :Get Sampling error:
###### Sam,pling Error = std dev / n^0.5
#
# step 3 : Get Margin of error:
###### Margin of Error = std dev  * sampling error





#################################
# find 90% conf interval 
mean <- 507
std_dev = 63
#Sample sizes
# a = 30
#b = 60
# c = what does increase sample size have on precision of estimates
#
# We need sampling error
# we need z score.  for 90% we look up 5%
#
#Part A
N_a <- 30
#Get sampling error
#sampling error is the std dev / N^0.5
#sampling_error = 39/35^0.5
Z_score <- qnorm(0.05)
Z_score
paste("Z score is",Z_score)
ROUNDED_Z_Score <- round(Z_score,3)
paste("Rounded z score is",ROUNDED_Z_Score)
#
Sampling_Error_A <- std_dev / N_a^0.5
Sampling_Error_A
ROUNDED_Sampling_Error_A <- round(Sampling_Error_A,2)
paste("Rounded sampling error is",ROUNDED_Sampling_Error_A)
#
# Margin of Error = z score * sampling error
MarginOfError <- ROUNDED_Z_Score * ROUNDED_Sampling_Error_A
ROUNDED_margin_of_error <- round(MarginOfError,4)
paste("Margin of error for 30 is",ROUNDED_margin_of_error)

#
#Next we calc the upper and lower limits
#upper_margin_of_error <- 247.8 + (ROUNDED_Z_score_at5_pct*ROUNDED_sampling_error)

upper_limit_30 <- 507 + (ROUNDED_Z_Score * ROUNDED_Sampling_Error_A  )
upper_limit_30
ROUNDED_upper_limit_30 <- round(upper_limit_30,2)
paste("the upper limit of n= 30 is ",ROUNDED_upper_limit_30)
#
#Next the lower limit
lower_limit_30 <- 507 - (ROUNDED_Z_Score * ROUNDED_Sampling_Error_A  )
lower_limit_30
ROUNDED_lower_limit_30 <- round(lower_limit_30,2)
paste("The Rounded lower limit of n= 30 is ",ROUNDED_lower_limit_30)
```
