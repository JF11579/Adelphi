## Next authenticate twitter acct api

```python
auth= tweepy.OAuthHandler(api_key , api_key_secret)
```

```python
auth.set_access_token(access_token_ , access_token_secret)
```

create api instance

```python
api = tweepy.API(auth)
```

### Access our tweets

```python
public_tweets = api.home_timeline()
```

```python
print(public_tweets)
```

```python
for tweet in public_tweets:
  print(tweet.text)
```

### Who tweeted it?

```python
print(public_tweets[0].user.screen_name)
```

### export tweets into csv file

```python
public_tweets = api.home_timeline()
```

```python
columns = ['Time' ,'User' , 'Tweet' ]
```

```python
data = []
```

```python
for tweet in public_tweets:
  data.append([tweet.created_at , tweet.user.screen_name , tweet.text])
```

```python
print(data)
```

### convert to DF

```python
df = pd.DataFrame(data , columns=columns)
```

```python
print(df)
```

```python
df.to_csv('tweets.csv')
```
